```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.catplot(x="day",y="tip",kind="strip",data=data)
```

```python
sns.catplot(x="day",y="tip",kind="strip",data=data,jitter=0)
```

```python
sns.catplot(x="day",y="tip",kind="swarm",data=data)
```

```python
sns.catplot(x="day",y="tip",kind="swarm",hue="sex",data=data)
```

```python
import numpy as np
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.relplot(x="total_bill",y="tip",data=data,kind="scatter")
```

```python
sns.relplot(x="total_bill",y="tip",data=data,kind="line")
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="sex",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="sex",size="size",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="sex",size="size",sizes=(50,200),data=data)
```
